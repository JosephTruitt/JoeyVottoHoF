## Conclusion
Joey Votto is no sure-fire Hall of Famer. He has apparent weaknesses in his resume. These include little post-season success or appearances due to being on a poor franchise most of his career. This is less important than players of other sports, as a single player in baseball has much less influence in a game than, for example, a QB in football or a player in basketball, but it is essential to note. Another weakness is his lower-than-average batting average and hits. The new wave of baseball analytics largely ignores these stats, but the Hall of Fame voters today likely will consider it when voting. Regardless, his elite OBP and good OPS should counter it. Lastly, Votto's inadequate defensive ability is something to note. He was not miserable, but definitely below average. He accumulated a -7.0 defensive WAR over the length of his career. Defensive ability, in general, is much less critical than offensive in baseball, particularly at first base, but it is important to note.

Votto's resume does include many strengths. The strongest being his OPS and WAR. OPS is the standard today to analyze a player's hitting ability, and for Votto to post a great stat indicates that he belongs in the Hall of Fame. Next, Votto's WAR does similar things. The fact he was able to have above median WAR while having a much below median career length does wonders for Votto's resume and maybe the accomplishment to get him in the Hall of Fame. 

Looking at Votto's resume, he belongs in the Hall of Fame, and he will make it there. It is important to remember that he does not have to be better or even better than slightly worse to make the Hall of Fame. You have to showcase that he belongs. With that being said, Joey Votto has a resume that would place him as an average hall of famer. His lack of post-season play offsets his better-than-average production, giving him a resume that would put him at average Hall of Fame status.
